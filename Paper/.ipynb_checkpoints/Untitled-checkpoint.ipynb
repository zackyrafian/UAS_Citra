```python
import cv2
import numpy as np
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

def convert_rgb_to_hsv(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

def compress_image(image, quality):
    buffer = BytesIO()
    image.save(buffer, format="JPEG", quality=quality)
    compressed_image = Image.open(buffer)
    return np.array(compressed_image)

def calculate_metrics(original, compressed):
    original_gray = cv2.cvtColor(original, cv2.COLOR_RGB2GRAY)
    compressed_gray = cv2.cvtColor(compressed, cv2.COLOR_RGB2GRAY)
    psnr_value = psnr(original_gray, compressed_gray)
    ssim_value = ssim(original_gray, compressed_gray)
    return psnr_value, ssim_value

# Contoh penggunaan
image_path = 'Gambar Sample/semangka.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Konversi ke HSV
image_hsv = convert_rgb_to_hsv(image_rgb)

# Kompresi dengan kualitas 50%
compressed_rgb = compress_image(Image.fromarray(image_rgb), 50)
compressed_hsv = compress_image(Image.fromarray(image_hsv), 50)

# Evaluasi kualitas
psnr_rgb, ssim_rgb = calculate_metrics(image_rgb, compressed_rgb)
psnr_hsv, ssim_hsv = calculate_metrics(image_rgb, cv2.cvtColor(compressed_hsv, cv2.COLOR_HSV2RGB))

print(f"PSNR RGB: {psnr_rgb}, SSIM RGB: {ssim_rgb}")
print(f"PSNR HSV: {psnr_hsv}, SSIM HSV: {ssim_hsv}")
```

```python
import cv2
import numpy as np
from PIL import Image
import os

def convert_rgb_to_hsv(image_rgb):
    return cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

def compress_image(image, quality, output_path):
    image.save(output_path, format="JPEG", quality=quality)
    return os.path.getsize(output_path)

def calculate_compression_ratio(original_size, compressed_size):
    return original_size / compressed_size

# Load image
image_path = 'Gambar Sample/sapi.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = convert_rgb_to_hsv(image_rgb)

# Convert to PIL format for compression
image_rgb_pil = Image.fromarray(image_rgb)
image_hsv_pil = Image.fromarray(image_hsv)

# Get original size in bytes
original_size = os.path.getsize(image_path)
print(f"Original size: {original_size} bytes")

# Define compression quality levels
compression_qualities = [10, 20, 30, 40, 50]

# Temporary file paths
temp_rgb_path = 'temp_rgb.jpg'
temp_hsv_path = 'temp_hsv.jpg'

# Store results
compression_results = {
    "quality": [],
    "rgb_size": [],
    "hsv_size": [],
    "rgb_ratio": [],
    "hsv_ratio": []
}

# Perform compression and calculate ratios
for quality in compression_qualities:
    rgb_size = compress_image(image_rgb_pil, quality, temp_rgb_path)
    hsv_size = compress_image(image_hsv_pil, quality, temp_hsv_path)
    
    rgb_ratio = calculate_compression_ratio(original_size, rgb_size)
    hsv_ratio = calculate_compression_ratio(original_size, hsv_size)
    
    compression_results["quality"].append(quality)
    compression_results["rgb_size"].append(rgb_size)
    compression_results["hsv_size"].append(hsv_size)
    compression_results["rgb_ratio"].append(rgb_ratio)
    compression_results["hsv_ratio"].append(hsv_ratio)

# Clean up temporary files
os.remove(temp_rgb_path)
os.remove(temp_hsv_path)

# Print results in a table format
import pandas as pd

df = pd.DataFrame(compression_results)
print(df)
```

```python
import cv2
import numpy as np
from PIL import Image
import os
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
import matplotlib.pyplot as plt
import pandas as pd

def convert_rgb_to_hsv(image_rgb):
    return cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

def compress_image(image, quality, output_path):
    image.save(output_path, format="JPEG", quality=quality)
    return Image.open(output_path), os.path.getsize(output_path)

def calculate_compression_ratio(original_size, compressed_size):
    return original_size / compressed_size

# Load image
image_path = 'Gambar Sample/sapi.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = convert_rgb_to_hsv(image_rgb)

# Convert to PIL format for compression
image_rgb_pil = Image.fromarray(image_rgb)
image_hsv_pil = Image.fromarray(image_hsv)

# Get original size in bytes
original_size = os.path.getsize(image_path)
print(f"Original size: {original_size} bytes")

# Define compression quality levels
compression_qualities = [10, 20, 30, 40, 50]

# Temporary file paths
temp_rgb_path = 'temp_rgb.jpg'
temp_hsv_path = 'temp_hsv.jpg'

# Store results
compression_results = {
    "quality": [],
    "rgb_size": [],
    "hsv_size": [],
    "rgb_ratio": [],
    "hsv_ratio": [],
    "rgb_psnr": [],
    "hsv_psnr": [],
    "rgb_ssim": [],
    "hsv_ssim": []
}

# Perform compression and calculate ratios
for quality in compression_qualities:
    compressed_rgb_image, rgb_size = compress_image(image_rgb_pil, quality, temp_rgb_path)
    compressed_hsv_image, hsv_size = compress_image(image_hsv_pil, quality, temp_hsv_path)
    
    rgb_ratio = calculate_compression_ratio(original_size, rgb_size)
    hsv_ratio = calculate_compression_ratio(original_size, hsv_size)
    
    # Convert back to numpy array for PSNR and SSIM calculation
    compressed_rgb_array = np.array(compressed_rgb_image)
    compressed_hsv_array = np.array(compressed_hsv_image)
    
    rgb_psnr = peak_signal_noise_ratio(image_rgb, compressed_rgb_array)
    hsv_psnr = peak_signal_noise_ratio(image_hsv, compressed_hsv_array)
    
    rgb_ssim = structural_similarity(image_rgb, compressed_rgb_array, multichannel=True, win_size=7, channel_axis=2)
    hsv_ssim = structural_similarity(image_hsv, compressed_hsv_array, multichannel=True, win_size=7, channel_axis=2)
    
    compression_results["quality"].append(quality)
    compression_results["rgb_size"].append(rgb_size)
    compression_results["hsv_size"].append(hsv_size)
    compression_results["rgb_ratio"].append(rgb_ratio)
    compression_results["hsv_ratio"].append(hsv_ratio)
    compression_results["rgb_psnr"].append(rgb_psnr)
    compression_results["hsv_psnr"].append(hsv_psnr)
    compression_results["rgb_ssim"].append(rgb_ssim)
    compression_results["hsv_ssim"].append(hsv_ssim)
    
    print(f"Quality: {quality}%")
    print(f"RGB compressed size: {rgb_size} bytes, Compression ratio: {rgb_ratio:.2f}, PSNR: {rgb_psnr:.2f}, SSIM: {rgb_ssim:.4f}")
    print(f"HSV compressed size: {hsv_size} bytes, Compression ratio: {hsv_ratio:.2f}, PSNR: {hsv_psnr:.2f}, SSIM: {hsv_ssim:.4f}\n")

# Clean up temporary files
os.remove(temp_rgb_path)
os.remove(temp_hsv_path)

# Print results in a table format
df = pd.DataFrame(compression_results)
print(df)

# Plot and save the compression ratio graph
plt.figure()
plt.plot(compression_results["quality"], compression_results["rgb_ratio"], label='RGB Compression Ratio')
plt.plot(compression_results["quality"], compression_results["hsv_ratio"], label='HSV Compression Ratio')
plt.xlabel('Quality Level')
plt.ylabel('Compression Ratio')
plt.legend()
plt.title('Comparison of Compression Ratios')
plt.savefig('Gambar Sample/sapi compression_ratio.png')

# Plot and save the PSNR graph
plt.figure()
plt.plot(compression_results["quality"], compression_results["rgb_psnr"], label='RGB PSNR')
plt.plot(compression_results["quality"], compression_results["hsv_psnr"], label='HSV PSNR')
plt.xlabel('Quality Level')
plt.ylabel('PSNR')
plt.legend()
plt.title('Comparison of PSNR')
plt.savefig('Gambar Sample/sapi psnr_values.png')

# Plot and save the SSIM graph
plt.figure()
plt.plot(compression_results["quality"], compression_results["rgb_ssim"], label='RGB SSIM')
plt.plot(compression_results["quality"], compression_results["hsv_ssim"], label='HSV SSIM')
plt.xlabel('Quality Level')
plt.ylabel('SSIM')
plt.legend()
plt.title('Comparison of SSIM')
plt.savefig('Gamabar Sample/sapi ssim_values.png')
```
